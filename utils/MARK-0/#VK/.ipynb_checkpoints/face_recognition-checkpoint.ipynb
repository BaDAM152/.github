```python
# Import required libraries
import face_recognition
import cv2
import numpy as np
import os
from sklearn.neighbors import KNeighborsClassifier
import matplotlib.pyplot as plt
%matplotlib inline

# Function to load and encode faces from training directory
def load_training_data(train_dir):
    face_encodings = []
    face_names = []
    
    # Loop through each person in the training directory
    for person_name in os.listdir(train_dir):
        person_dir = os.path.join(train_dir, person_name)
        if os.path.isdir(person_dir):
            # Loop through each training image for the current person
            for img_name in os.listdir(person_dir):
                img_path = os.path.join(person_dir, img_name)
                # Load and encode face
                image = face_recognition.load_image_file(img_path)
                face_locations = face_recognition.face_locations(image)
                
                if len(face_locations) > 0:
                    face_encoding = face_recognition.face_encodings(image, face_locations)[0]
                    face_encodings.append(face_encoding)
                    face_names.append(person_name)
                    
    return face_encodings, face_names

# Load training data
train_dir = "data/train"
face_encodings, face_names = load_training_data(train_dir)

# Train KNN classifier
knn = KNeighborsClassifier(n_neighbors=1, metric='euclidean')
knn.fit(face_encodings, face_names)

# Function to recognize faces in an image
def recognize_faces(image_path, knn_clf, threshold=0.5):  # Lowered threshold for better multi-face detection
    # Load test image
    image = face_recognition.load_image_file(image_path)
    
    # Find all faces in the image
    face_locations = face_recognition.face_locations(image)
    face_encodings = face_recognition.face_encodings(image, face_locations)
    
    # Convert image from RGB to BGR for OpenCV
    image = cv2.cvtColor(image, cv2.COLOR_RGB2BGR)
    
    # Store recognition results
    recognition_results = []
    
    # Loop through each face in the test image
    for (top, right, bottom, left), face_encoding in zip(face_locations, face_encodings):
        # Get the distances and predictions
        distances, indices = knn_clf.kneighbors([face_encoding])
        closest_distance = distances[0][0]
        
        # Convert distance to confidence score (0-100%)
        confidence_score = (1 - closest_distance) * 100
        
        # Predict the name using KNN classifier
        predicted_name = knn_clf.predict([face_encoding])[0]
        
        # If confidence is too low, mark as Unknown
        if confidence_score < threshold * 100:
            predicted_name = "Unknown"
            
        # Store results
        recognition_results.append((predicted_name, confidence_score))
            
        # Draw rectangle around the face
        cv2.rectangle(image, (left, top), (right, bottom), (0, 255, 0), 2)
        
        # Draw name on top of the box with confidence
        top_label = f"{predicted_name} ({confidence_score:.1f}%)"
        label_size = cv2.getTextSize(top_label, cv2.FONT_HERSHEY_DUPLEX, 0.6, 1)[0]
        cv2.rectangle(image, (left, top - 35), (left + label_size[0] + 20, top), (0, 255, 0), cv2.FILLED)
        font = cv2.FONT_HERSHEY_DUPLEX
        cv2.putText(image, top_label, (left + 6, top - 6), font, 1.0, (10, 10, 10), 1)
    
    # Convert back to RGB for matplotlib display
    image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
    return image, recognition_results

# Test the model on a single image
test_image_path = "data/test/test10.jpeg"  # Make sure this image contains multiple faces
result_image, recognition_results = recognize_faces(test_image_path, knn)

# Create figure with two subplots
plt.figure(figsize=(12, 10))

# Display the image
plt.subplot(2, 1, 1)
plt.imshow(result_image)
plt.axis('off')
plt.title("Face Recognition Results", pad=20)

# Display recognition results as text
plt.subplot(2, 1, 2)
plt.axis('off')
text_output = "People Identified in the Image:\n\n"

# Group results by person
identified_people = {}
for name, confidence in recognition_results:
    if name in identified_people:
        identified_people[name].append(confidence)
    else:
        identified_people[name] = [confidence]

# Display results grouped by person
for name, confidences in identified_people.items():
    avg_confidence = sum(confidences) / len(confidences)
    count = len(confidences)
    text_output += f"{name}: {count} {'occurrence' if count == 1 else 'occurrences'} "
    text_output += f"(Avg. Confidence: {avg_confidence:.2f}%)\n"

text_output += f"\nTotal people detected: {len(recognition_results)}"
text_output += f"\nModel trained on {len(face_names)} images of {len(set(face_names))} person(s)"
plt.text(0.1, 0.5, text_output, fontsize=12, verticalalignment='center')

plt.tight_layout()
plt.show()

# Print summary to console
print("\nRecognition Results:")
for name, confidences in identified_people.items():
    avg_confidence = sum(confidences) / len(confidences)
    count = len(confidences)
    print(f"{name}: {count} {'occurrence' if count == 1 else 'occurrences'} "
          f"(Avg. Confidence: {avg_confidence:.2f}%)")
print(f"\nTotal people detected: {len(recognition_results)}")
print(f"Model trained on {len(face_names)} images of {len(set(face_names))} person(s)")
```
